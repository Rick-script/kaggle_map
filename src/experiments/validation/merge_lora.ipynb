```python
import os
from safetensors.torch import load_file
from safetensors.torch import save_file
```

```python
lora_dir = "exp022_qwen2.5-14b-lora-fold0/output/model/adapter_model.safetensors"
d1 = load_file(lora_dir)
lora_dir = "exp022_qwen2.5-14b-lora-fold1/output/model/adapter_model.safetensors"
d2 = load_file(lora_dir)
lora_dir = "exp022_qwen2.5-14b-lora-fold2/output/model/adapter_model.safetensors"
d3 = load_file(lora_dir)
lora_dir = "exp022_qwen2.5-14b-lora-fold3/output/model/adapter_model.safetensors"
d4 = load_file(lora_dir)
lora_dir = "exp022_qwen2.5-14b-lora-fold4/output/model/adapter_model.safetensors"
d5 = load_file(lora_dir)
```

```python
d = {}
for k, v in d1.items():
    v = d1[k] + d2[k] + d3[k] + d4[k] + d5[k]
    v = v / 5.
    d[k] = v

save_path = "validation/exp022_qwen2.5-14b-lora-merge/peft"
os.makedirs(save_path, exist_ok=True)

save_file(d, f"{save_path}/adapter_model.safetensors")
print('merge success!!!')
```

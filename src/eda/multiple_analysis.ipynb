```python
# データの読み込み
train = pd.read_csv("../../kaggle/input/map-charting-student-math-misunderstandings/train.csv")
print(train.shape)
train.head()
```

## データの確認

```python
# QuestionIdごとの選択肢マッピングを定義
question_choices = {
    31772: {r'\( \frac{1}{3} \)': 'A', r'\( \frac{3}{6} \)': 'B', r'\( \frac{3}{8} \)': 'C', r'\( \frac{3}{9} \)': 'D'},
    31774: {r'\( \frac{1}{12} \)': 'A', r'\( 3 \)': 'B', r'\( \frac{1}{3} \)': 'C', r'\( \frac{6}{2} \)': 'D'},
    31777: {r'\( 72 \)': 'A', r'\( 24 \)': 'B', r'\( 60 \)': 'C', r'\( 48 \)': 'D'},
    31778: {r'\( 3 \)': 'A', r'\( 4 \)': 'B', r'\( 6 \)': 'C', r'\( 9 \)': 'D'},
    32829: {r'\( 4 \)': 'A', r'\( 12 \)': 'B', r'\( 48 \)': 'C', r'\( 22 \)': 'D'},
    32833: {r'\( \frac{10}{15} \)': 'A', r'\( \frac{2}{15} \)': 'B', r'\( 5 \frac{2}{3} \)': 'C', r'\( 3 \frac{1}{3} \)': 'D'},
    32835: {r'\( 6 \)': 'A', r'\( 6.2 \)': 'B', r'\( 6.079 \)': 'C', r'\( 6.0001 \)': 'D'},
    33471: {r'\( 8 \)': 'A', r'\( 9 \)': 'B', r'\( 15 \)': 'C', r'\( 3 \)': 'D'},
    33472: {r'\( \frac{3}{8} \)': 'A', r'\( \frac{3}{15} \)': 'B', r'\( \frac{11}{30} \)': 'C', r'\( \frac{11}{15} \)': 'D'},
    33474: {r'\( \frac{1}{3}+\frac{2}{3} \)': 'A', r'\( \frac{1}{3} \times \frac{2}{3} \)': 'B', r'\( \frac{2}{3} \div \frac{1}{3} \)': 'C', r'\( \frac{2}{3}-\frac{1}{3} \)': 'D'},
    76870: {r'\( 10 \)': 'A', r'\( 5 \)': 'B', r'\( 6 \)': 'C', 'Not enough information': 'D'},
    89443: {r'\( -13 \)': 'A', r'\( 13 \)': 'B', r'\( 3 \)': 'C', r'\( -3 \)': 'D'},
    91695: {r'\( 26 \)': 'A', r'\( 20 \)': 'B', r'\( 36 \)': 'C', r'\( 22 \)': 'D'},
    104665: {r'\( 768 \) hours': 'A', r'\( 48 \) hours': 'B', r'\( 192 \) hours': 'C', r'\( 64 \) hours': 'D'},
    109465: {'Certain': 'A', 'Likely': 'B', 'Unlikely': 'C', 'Impossible': 'D'}
}

# データに選択肢列を追加
def get_choice(row):
    mc_answer = row['MC_Answer']
    question_id = row['QuestionId']
    return question_choices[question_id].get(mc_answer, 'Unknown')

train['Choice'] = train.apply(get_choice, axis=1)
```

```python
# 選択肢ごとのカテゴリ比率を計算
choice_category_counts = train.groupby(['Choice', 'Category']).size().unstack(fill_value=0)
choice_category_ratios = choice_category_counts.div(choice_category_counts.sum(axis=1), axis=0) * 100

print("選択肢ごとのカテゴリ比率 (%):")
display(choice_category_ratios.round(2))
```

```python
# 選択肢ごとのカテゴリ比率を計算
choice_category_counts = train.groupby(['QuestionId', 'Choice', 'Category']).size().unstack(fill_value=0)
choice_category_ratios = choice_category_counts.div(choice_category_counts.sum(axis=1), axis=0) * 100

print("選択肢ごとのカテゴリ比率 (%):")
display(choice_category_ratios.round(2))
```
